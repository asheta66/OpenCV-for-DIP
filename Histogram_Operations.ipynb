**What is a histogram?**

**A histogram in image processing represents the distribution of pixel intensities in an image. It provides a visual representation of how many pixels in the image have specific intensity values. Histograms are commonly used for various image analysis tasks, including contrast adjustment, thresholding, and equalization.**

```python
import cv2
import matplotlib.pyplot as plt

# Load the Lenna image
image_path = 'Lenna.png'
lenna = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Calculate the histogram
histogram, bins = np.histogram(lenna.flatten(), bins=256, range=[0,256])

# Plot the histogram
plt.figure(figsize=(6, 4))
plt.plot(histogram, color='black')
plt.title('Histogram for Lenna Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**This histogram shows the distribution of pixel intensities in the Lenna image. The x-axis represents the pixel intensity values, and the y-axis represents the frequency of pixels with those intensity values. Analyzing the histogram can provide insights into the image's overall brightness and contrast.**
