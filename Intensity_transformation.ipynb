**Intensity transformation, also known as contrast enhancement or intensity mapping, is a fundamental technique in image processing that aims to adjust the pixel intensities in an image. **

image.png

```python
import cv2
import matplotlib.pyplot as plt

# Load the Lenna image
image_path = 'Lenna.png'
lenna = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Linear Contrast Stretching (Normalization)
min_val = lenna.min()
max_val = lenna.max()
new_min = 0
new_max = 255

# Linear Contrast Stretching Transformation
stretched_lenna = ((lenna - min_val) / (max_val - min_val)) * (new_max - new_min) + new_min

# Convert the stretched image to uint8 (8-bit) format
stretched_lenna = stretched_lenna.astype('uint8')

# Display original and stretched images side by side
plt.figure(figsize=(5, 5))

plt.subplot(1, 2, 1)
plt.imshow(lenna, cmap='gray')
plt.title('Original Lenna')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched_lenna, cmap='gray')
plt.title('Linear Contrast Stretched Lenna')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Lenna image
image_path = 'Lenna.png'
lenna = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Logarithmic Transformation
c = 1  # Constant for scaling
log_transformed_lenna = c * np.log1p(lenna)

# Convert the log-transformed image to uint8 (8-bit) format
log_transformed_lenna = log_transformed_lenna.astype('uint8')

# Display original and log-transformed images side by side
plt.figure(figsize=(5, 5))

plt.subplot(1, 2, 1)
plt.imshow(lenna, cmap='gray')
plt.title('Original Lenna')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(log_transformed_lenna, cmap='gray')
plt.title('Logarithmic Transformed Lenna')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Lenna image
image_path = 'Lenna.png'
lenna = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Power-Law (Gamma) Transformation
c = 1  # Constant for scaling
gamma = 0.5  # Gamma value for the transformation
gamma_transformed_lenna = c * np.power(lenna, gamma)

# Convert the gamma-transformed image to uint8 (8-bit) format
gamma_transformed_lenna = gamma_transformed_lenna.astype('uint8')

# Display original and gamma-transformed images side by side
plt.figure(figsize=(5, 5))

plt.subplot(1, 2, 1)
plt.imshow(lenna, cmap='gray')
plt.title('Original Lenna')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_transformed_lenna, cmap='gray')
plt.title('Gamma Transformed Lenna (Gamma = 0.5)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Lenna image
image_path = 'Lenna.png'
lenna = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define piecewise linear transformation function
def piecewise_linear_transform(image, breakpoints, slopes):
    result = np.piecewise(image, [image < breakpoints[0], (image >= breakpoints[0]) & (image < breakpoints[1]), image >= breakpoints[1]],
                          [lambda x: slopes[0] * x,
                           lambda x: slopes[1] * (x - breakpoints[0]) + slopes[0] * breakpoints[0],
                           lambda x: slopes[2] * (x - breakpoints[1]) + slopes[1] * (breakpoints[1] - breakpoints[0]) + slopes[0] * breakpoints[0]])
    return np.clip(result, 0, 255).astype(np.uint8)

# Define piecewise linear transformation parameters
breakpoints = [80, 160]
slopes = [2, 1, 2]

# Apply piecewise linear transformation
piecewise_transformed_lenna = piecewise_linear_transform(lenna, breakpoints, slopes)

# Display original and piecewise transformed images side by side
plt.figure(figsize=(6, 6))

plt.subplot(1, 2, 1)
plt.imshow(lenna, cmap='gray')
plt.title('Original Lenna')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(piecewise_transformed_lenna, cmap='gray')
plt.title('Piecewise Linear Transformed Lenna')
plt.axis('off')

plt.tight_layout()
plt.show()
```
